# Exercise 11.1

### Linear fit

$$
f(x)=2x+1
$$

These are the parameters that define the line:

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

And these are the parameters we can change in order to improve the neural network performance:

```python
n_train = 1000
sigma = 0.8 # noise standard deviation
n_epochs = 30
```

Let's generate the **training set** and the **validation set**.

```python
import numpy as np

n_valid = int(n_train/10)   # tipically the majority of the data is partitioned into the training set.

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, n_train)
x_valid = np.random.uniform(-1, 1, n_valid)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

y_train = np.random.normal(m * x_train + b, sigma) 
y_valid = np.random.normal(m * x_valid + b, sigma)
```

Plots of the training and validation set compared with line described by the equation above.

```python
import matplotlib.pyplot as plt

f, axarr = plt.subplots(1,2)

axarr[0].plot(x_valid, y_target)
axarr[0].scatter(x_train, y_train, color='r')
axarr[0].grid(True); 
axarr[0].set_title('Training set')

axarr[1].plot(x_valid, y_target)
axarr[1].scatter(x_valid, y_valid, color='r')
axarr[1].grid(True); 
axarr[1].set_title('Validation set')

plt.show()
```

Let's create a very simple network with one hidden layer composed of only one neuron. In the following no activation is used.

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

Let's train the network feeding it with the training set. Using the Stochastic Gradient Descent the network will find the weights that minimize the loss function (mse).

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=n_epochs,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

Changing the parameters n_train, sigma and n_epochs it is evident that the fundamental parameter for convergence is **n_train**. The noise could be very high, but if the training set is sufficiently wide, the network identify the line. On the contrary if the training set is poor, we can use a high number of epochs and a tiny noise, without reaching convergence.

# Exercise 11.2

### Polinomial fit

$$
f(x)=3x^3-2x^2-3x+4
$$

Let's try to fit a polynomial of order 3 for $x \in [-1,1]$.

```python
# f(x) = ax^3 + bx^2 + cx + d
a = 3
b = -2
c = -3
d = 4

def f(x):
        return a*x**3 + b*x**2 + c*x + d
```

```python
n_train = 5000
sigma = 0.4 # noise standard deviation
n_epochs = 30
```

Let's generate the **training set** and the **validation set**.

```python
import numpy as np

n_valid = int(n_train/10)   # tipically the majority of the data is partitioned into the training set.

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, n_train)
x_valid = np.random.uniform(-1, 1, n_valid)
x_valid.sort()
y_target = f(x_valid) # ideal (target) linear function

y_train = np.random.normal(f(x_train), sigma) 
y_valid = np.random.normal(f(x_valid), sigma)
```

```python
import matplotlib.pyplot as plt

f, axarr = plt.subplots(1,2)

axarr[0].plot(x_valid, y_target)
axarr[0].scatter(x_train, y_train, color='r')
axarr[0].grid(True); 
axarr[0].set_title('Training set')

axarr[1].plot(x_valid, y_target)
axarr[1].scatter(x_valid, y_valid, color='r')
axarr[1].grid(True); 
axarr[1].set_title('Validation set')

plt.show()
```

Some of the most famous optimizers are:

```python
from tensorflow import keras

sgd = keras.optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
adag = keras.optimizers.Adagrad(lr=0.01, epsilon=None, decay=0.0)
adad = keras.optimizers.Adadelta(lr=1.0, rho=0.95, epsilon=None, decay=0.0)
```

### Trial 1

Let's begin from the simplest case: we build a network with one hidden layer composed of 2 neurons without activation.

```python
import tensorflow as tf


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model_1 = tf.keras.Sequential()
model_1.add(Dense(2, input_shape=(1,)))
model_1.add(Dense(1))

# compile the model choosing optimizer, loss and metrics objects
model_1.compile(optimizer=sgd, loss='mse', metrics=['mse'])
```

```python
model_1.summary()
```

Network training:

```python
history = model_1.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=n_epochs,
          shuffle=True, 
          validation_data=(x_valid, y_valid))
```

```python
model_1.get_weights()
```

Once the training process has completed, we can evaluate the model over different test dataset. The evaluate method returns loss and accuracy.

```python
# evaluate model
score = model_1.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model_1.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

This network is too simple to fit a polynomial of order 3. We try to add neurons and activations.

### Trial 2

We increment the number of neurons but we consider only one hidden layer. It is also needed to introduce some non linearity effect. We use a typical activation function: ReLu.

```python
model_2 = tf.keras.Sequential()
model_2.add(Dense(30, input_shape=(1,), activation='relu'))
model_2.add(Dense(1))

# compile the model choosing optimizer, loss and metrics objects
model_2.compile(optimizer=sgd, loss='mse', metrics=['mse'])

history = model_2.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=n_epochs,
          shuffle=True, 
          validation_data=(x_valid, y_valid))
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model_2.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

### Trial 3
We can do better. Increase the number of layers

```python
n_epochs=50

model_3 = tf.keras.Sequential()
model_3.add(Dense(10, input_shape=(1,)))
model_3.add(Dense(10, activation='relu'))
model_3.add(Dense(10, activation='tanh'))
model_3.add(Dense(1, activation='relu'))

# compile the model choosing optimizer, loss and metrics objects
model_3.compile(optimizer=adag, loss='mse', metrics=['mse'])

history = model_3.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=n_epochs,
          shuffle=True, 
          validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score = model_3.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model_3.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

Not bad. To improve our fit we used a tanh activation function in the second hidden layer in order to smoothen the fit.

The result is satisfying for each of the optimizers listed above. Here we used Adagrad, which seems to produce a slightly better result. Let's test the network on a new data set that includes data outside the range $[-1,1]$.

```python
x = np.random.uniform(-1.5, 1.5, n_valid)
x.sort()

x_predicted = np.random.uniform(-1.5, 1.5, 100)
y_predicted = model_3.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x, a*x**3 + b*x**2 + c*x + d)
plt.grid(True)
plt.show()
```

We notice that the network is able to predict only in the range in which the training took place.

If we increase the number of neurons in each layer, the network becomes incredibly powerful. But pay attention: the network is overfitting the data. For this reason we tried to work with a little number of neurions per layer.

# Exercise 11.1

### Trigonometric 2D function fit
$$f(x,y) = \sin(x_1^2+x_2^2)$$

in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.



```python
import numpy as np

def f(x1,x2):
        return np.sin(x1**2+x2**2)
```

```python
n_train = 5000
sigma = 0.4 # noise standard deviation
n_epochs = 30
```

Generate training and validation sets.

```python
n_valid = int(n_train/10)   # tipically the majority of the data is partitioned into the training set.

# generate training inputs
np.random.seed(0)

x_train = np.zeros((n_train, 2))
x_valid = np.zeros((n_valid, 2))

x_train = np.random.uniform(-1.5, 1.5, (n_train, 2))
x_valid = np.random.uniform(-1.5, 1.5, (n_valid, 2))

#x_valid.sort()


y_target = f(x_valid[:,0], x_valid[:,1])

y_train = np.random.normal(f(x_train[:,0], x_train[:,1]), sigma) 
y_valid = np.random.normal(f(x_valid[:,0], x_valid[:,1]), sigma)


```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D

x1_train = x_train[:,0]
x2_train = x_train[:,1]
x1_valid = x_valid[:,0]
x2_valid = x_valid[:,1]



# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=(18,4))#plt.figaspect(0.5))

#===============
#  First subplot
#===============
# set up the axes for the first plot
ax = fig.add_subplot(1, 3, 1, projection='3d')
ax.scatter(x1_train, x2_train, y_train, c=y_train, marker='.')
ax.set_title('Training set')


#===============
# Second subplot
#===============
# set up the axes for the second plot
ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.scatter(x1_valid, x2_valid, y_valid, c=y_valid, marker='.')
ax.set_title('Validation set')

#ax.view_init(20, 30)

#===============
# Third subplot
#===============
# set up the axes for the third plot
ax = fig.add_subplot(1, 3, 3, projection='3d')
X = np.arange(-1.5, 1.5, 0.01)
Y = np.arange(-1.5, 1.5, 0.01)
X, Y = np.meshgrid(X, Y)
Z = np.sin(X**2 + Y**2)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('Target function')

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)


plt.show()
```

```python
n_epochs=30

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(30, input_shape=(2,)))
model.add(Dense(30, activation='relu'))
model.add(Dense(30, activation='relu'))
#model.add(Flatten())
model.add(Dense(1, activation='relu'))


# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])


history = model.fit(x=x_train, y=y_train,
          batch_size=32, epochs=n_epochs,
          shuffle=True, 
          validation_data=(x_valid, y_valid))
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1.5, 1.5, (1000, 2))
y_predicted = model.predict(x_predicted)


fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(x_predicted[:,0], x_predicted[:,1], y_predicted, marker='*')

X = np.arange(-1.5, 1.5, 0.01)
Y = np.arange(-1.5, 1.5, 0.01)
X, Y = np.meshgrid(X, Y)
Z = np.sin(X**2 + Y**2)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_title('Target function')

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
#fig.colorbar(surf, shrink=0.5, aspect=5)


plt.show()
```
