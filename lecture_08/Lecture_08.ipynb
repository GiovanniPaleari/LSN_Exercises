# Exercise 08.1 & 08.2

We have to variationally optimize the ground state of a single quantum particle 1D confined by the external potential $ V(x) = x^4 - \frac{5}{2}x^2$. The Variational Monte Carlo method uses a trial wave function, parametrized by a set of variational parameters. An approximate model for the ground state wave function is the following trial wave function:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$
We have to find the variational parameters $\sigma$ and $\mu$ that minimize the energy of the state.

We used a Metropolis Monte Carlo code which has the very same structure of the Monte_Carlo_NVT code. We want to sample the probability distribution $$p(x)=\frac{|\Psi_T(x)|^2}{\int dx|\Psi_T(x)|^2}$$ in order to calculate the energy integral $$
\langle {\hat H} \rangle_T = 
\int dx  \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} p(x)
$$ The acceptance probability of a uniform move of fixed step $x \to y$ is given by $A(x|y)=\frac{|\Psi_T^{\sigma,\mu}(x)|^2}{|\Psi_T^{\sigma,\mu}(y)|^2}$. Note that this is exactly what we implemented in our first Metropolis MC code, where we wanted to sample the hydrogen wave functions.

Given the parameters $\sigma$ and $\mu$ and the amplitude of a single step, the program is able to compute the energy of the corrisponding quantum state. At each MC step the program accumulates values of $$\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}=V(x)-\frac{1}{2 \Psi_T(x)} \frac{\partial^2}{\partial x^2}\Psi_T(x)$$ where we put $\hbar=1$. If we are sampling the correct probability distribution the sum of these values properly rescaled is a good estimation of the integral $\langle {\hat H} \rangle_T$.

Since we are not so good in optimizing problems for now, let's use a *brute force* method. We generate a grid of possible values for $\sigma$ and $\mu$ and we execute the program many times with the different parameters. At the end of each execution we compare the energy obtained with the lowest energy obtained so far. If it is bigger, forget about it! But if it is lower, we save it as the new best approximated energy for the ground state. At the end of this process, we will have the best parameters $\sigma$ and $\mu$ of the grid. Our possible values of the variational parameters are:
$$\sigma \in [0.1,2.0] $$
$$\mu \in [0.1,2.0] $$

Let's plot the energy for the corresponding simulation compared to the energy obtained with the numerical solution that transforms the Schrodinger equation into a matrix equation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

iblk, e, err = np.loadtxt('Ex_08.1-2/Data/Energy_sigma0.6_mu0.8.txt',usecols=(0,1,2), delimiter='\t', unpack='true')

plt.figure(figsize=(16,8))

plt.errorbar(iblk, e, yerr=err)
plt.plot([0,100],[-0.46046587969271535,-0.46046587969271535])
plt.xlabel('Number of blocks')
plt.ylabel('Energy')
plt.show()
```

```python
def f(x, mu, sigma, c):
    return c*(np.exp(-0.5*((x-mu)/sigma)**2)+np.exp(-0.5*((x+mu)/sigma)**2))**2

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])

# Plot a few things
plt.figure(figsize=(8,5))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2, color='r', label='GS')

p_opt, p_cov = curve_fit(f, x, (psi[0])**2) #, bounds=([0,0,-1],[2,3,+3]))
y_fit = f(x,p_opt[0],p_opt[1],p_opt[2])
#plt.plot(x,y_fit, label='fit') # plotting fitted function
#print('Fit parameters:')
#print('mu = ', p_opt[0])
#print('sigma = ', p_opt[1])
print('c = ', p_opt[2])

plt.plot(x, f(x, 0.8, 0.6, 0.4), label='var_mc', color='g')

xbin, histo = np.loadtxt('Ex_08.1-2/histo.txt',usecols=(0,1), delimiter='\t', unpack='true')

plt.bar(xbin, histo, width=0.03, label='histogram var')

plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.legend()
plt.show()
```

In this picture we compare:
- the histogram representing the variationally optimized ground state wave function
- the analytic curve with parameters $\sigma=0.6$ and $\mu=0.8$
- the numerical solution obtained by transforming the Schrodinger equation into a matrix equation

The two approaches gives slightly different results. In particular the wave function obtained in my simulations has a bigger energy. This may be due to the *brute force* optimization technique and to the specific form of the trial wave function.
