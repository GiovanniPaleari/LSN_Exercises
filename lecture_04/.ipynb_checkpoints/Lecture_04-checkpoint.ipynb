# Exercise 04.1

The possibility to restart reading not only a actual spatial configuration $r(t)$ but also an old spatial configuration $r(t-dt)$ is of vital importance for the equilibration of our simulation.
In order to prevent superposition of particles we always choose a crystal FCC structure as the initial configuration of our Molecular Dynamics (MD) simulation. This means that at the very beginning of the simulation the system is near a minimum for the potential energy if we are simulating a liquid or a solid phase. A few iterations of the Verlet algorithm will drive the system away from this minimum increasing the potential energy. Since the Verlet algorithm has excellent energy-conserving properties, the kinetic energy and therefore the temperature will decrease. This is a problem, because we want to simulate the system at a fixed temperature, given by the user. But our system is isolated and conserves energy, not temperature and Verlet algorithm is excellent in implementing this conservation. We are trying to force our microcanonical system to work at a given temperature. So after few Verlet steps the temperature is lower than the initial one. We can exploit the restart option to restart the simulation with the desired temperature. We do this by rescaling the velocities in order to match the desired temperature. Basically we are restarting the simulation after heating up the system as necessary. We do this many times until the system is equilibrated around the target temperature. This is what we call the **equilibration** phase. For the gas we cannot say that the intial configuration is a minimum for the potential energy because density is very low. After few Verlet steps the temperature will grow up, and we will use the restart option in order to adjust it.

In order to understand what we just said, the figure below shows the equilibration phase for a liquid system characterized by the following parameters in Lennard-Jones units:
- density: $\rho^\star = 0.8$
- temperature: $T^\star = 1.1$ 
- cut-off radius: $r_c = 2.5$

We can see the downfalls of temperature compared to the target temperature and the heating up of the system in correspondence of the steps of the total energy. After 5 restarts of the simulation, we have reached the target temperature and we can start the measurements of physical quantities.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,5))
temp = np.loadtxt("Ex_04/Figures/Eq_output_temp.dat", unpack='true')
etot = np.loadtxt("Ex_04/Figures/Eq_output_etot.dat", unpack='true')

plt.plot(temp, label='Temperature')
plt.plot(etot, label='Total Energy')
plt.plot([1.,1500],[0.8,0.8], color='r', label='Target Temperature')
plt.xlabel('Number of steps (/10)')

plt.legend(loc='center right')

plt.show()
```

# Exercises 04.2 & 04.3

We upgraded the MD code including block averaging. We divided the $nstep$ of the simulation in $nblocks$ and coded the calculation of the average values and the uncertainties for $E/N$, $U/N$, $K/N$ and $T$. Then we included the calculation of the pressure $P$ which is estimated via the virial theorem as: $$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

# Exercise 04.4

Let's simulate a physical system! In particular we perform MD simulations via a Lennard-Jones model of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) and Krypton ($\sigma = 0.364$ nm, $\epsilon/k_B = 164$ K, $m=83.798$ amu) in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

After equilibrating the system we perform measurements on a simulation characterized by $nstep$ and $nblocks$.
We show pictures of the obtained average values and uncertainties for the total energy per particle, $E/N$, the potential energy per particle, $U/N$, the kinetic energy per particle, $K/N$, the pressure $P$ and the temperature, $T$ in SI units.

For the liquid phase: $nstep=10^4$ and $nblocks=20$.

```python
ave_temp, err_temp = np.loadtxt("Ex_04/Figures/Liq_ave_temp.out", usecols=(0,1), unpack='true')
ave_ekin, err_ekin = np.loadtxt("Ex_04/Figures/Liq_ave_ekin.out", usecols=(0,1), unpack='true')
ave_epot, err_epot = np.loadtxt("Ex_04/Figures/Liq_ave_epot.out", usecols=(0,1), unpack='true')
ave_etot, err_etot = np.loadtxt("Ex_04/Figures/Liq_ave_etot.out", usecols=(0,1), unpack='true')
ave_pr, err_pr = np.loadtxt("Ex_04/Figures/Liq_ave_pr.out", usecols=(0,1), unpack='true')

k_b=1.38*10**(-23)
eps_arg=120
eps_kr=164
sigma_arg=0.34
sigma_kr=0.364

#ARGON
ave_temp *= eps_arg
err_temp *= eps_arg
ave_ekin *= eps_arg*k_b
err_ekin *= eps_arg*k_b
ave_epot *= eps_arg*k_b
err_epot *= eps_arg*k_b
ave_etot *= eps_arg*k_b
err_etot *= eps_arg*k_b
ave_pr *= (eps_arg*k_b/sigma_arg**3)*10**(27)
err_pr *= (eps_arg*k_b/sigma_arg**3)*10**(27)

x = np.arange(0,20,1)

f, axarr = plt.subplots(5,1, figsize=(10,20))

axarr[0].errorbar(x,ave_temp,yerr=err_temp, color='b')
axarr[0].set(xlabel='Number of blocks')
axarr[0].set(ylabel='Temperature (K)')
axarr[0].set_title("Liquid Argon", fontsize=16)

axarr[1].errorbar(x,ave_ekin,yerr=err_ekin, color='r')
axarr[1].set(xlabel='Number of blocks')
axarr[1].set(ylabel='Kinetic Energy (J)')

axarr[2].errorbar(x,ave_epot,yerr=err_epot, color='g')
axarr[2].set(xlabel='Number of blocks')
axarr[2].set(ylabel='Potential Energy (J)')

axarr[3].errorbar(x,ave_etot,yerr=err_etot, color='c')
axarr[3].set(xlabel='Number of blocks')
axarr[3].set(ylabel='Total Energy (J)')

axarr[4].errorbar(x,ave_pr,yerr=err_pr, color='m')
axarr[4].set(xlabel='Number of blocks')
axarr[4].set(ylabel='Pressure (Pa)')

#plt.legend(loc='center right')

plt.show()
```

```python
#KRYPTON
ave_temp *= eps_kr/eps_arg
err_temp *= eps_kr/eps_arg
ave_ekin *= eps_kr/eps_arg
err_ekin *= eps_kr/eps_arg
ave_epot *= eps_kr/eps_arg
err_epot *= eps_kr/eps_arg
ave_etot *= eps_kr/eps_arg
err_etot *= eps_kr/eps_arg
ave_pr *= (eps_kr/eps_arg)*(sigma_arg/sigma_kr)**3
err_pr *= (eps_kr/eps_arg)/(sigma_arg/sigma_kr)**3

f, axarr = plt.subplots(5,1, figsize=(10,20))

axarr[0].errorbar(x,ave_temp,yerr=err_temp, color='b')
axarr[0].set(xlabel='Number of blocks')
axarr[0].set(ylabel='Temperature (K)')
axarr[0].set_title("Liquid Krypton", fontsize=16)

axarr[1].errorbar(x,ave_ekin,yerr=err_ekin, color='r')
axarr[1].set(xlabel='Number of blocks')
axarr[1].set(ylabel='Kinetic Energy (J)')

axarr[2].errorbar(x,ave_epot,yerr=err_epot, color='g')
axarr[2].set(xlabel='Number of blocks')
axarr[2].set(ylabel='Potential Energy (J)')

axarr[3].errorbar(x,ave_etot,yerr=err_etot, color='c')
axarr[3].set(xlabel='Number of blocks')
axarr[3].set(ylabel='Total Energy (J)')

axarr[4].errorbar(x,ave_pr,yerr=err_pr, color='m')
axarr[4].set(xlabel='Number of blocks')
axarr[4].set(ylabel='Pressure (Pa)')

#plt.legend(loc='center right')

plt.show()
```
